### Get the HTML from the website as a text file

```python
type(beirut_html)
```

```python
# now we have to catch the useful links in the website that links us to the page with lyrics
# we use 'Regular Expressions'
```

```python
name_of_artist = 'Beirut'
```

```python
pattern = '/lyric/\d+/'+name_of_artist+'/[\w+%-]+' 
```

```python
type(pattern)
```

```python
# now that the pattern was recognized, we are going to find that pattern in the 'beirut_html' and assign it to 'lyric_url'
lyric_url = re.findall(pattern, beirut_html) 
lyric_url 
```

```python
type(lyric_url)
```

```python
## looks like i am on the right way and found the lyric_url, BUT it does not appear like a real link
```

```python
## when i ran it the error was that it could not find the link. 
## something was missing in the beginning of the link- need to be added- 
## so i go back and add it to make it like a real internet link

modi_lyric_url =['https://www.lyrics.com/' + link for link in lyric_url]
modi_lyric_url[0:10]
```

```python
type(modi_lyric_url)
```

```python
# some testing to see if it can find one specific songlink --> eg. look into element number 2 it modi_lyric_url list!
songlink = modi_lyric_url[2]
response =requests.get(songlink)
songlink
```

```python
#nice! it got the url number3 --> the sound is exile--> it starts from 0 so element number 2 is actually the 3rd link
```

```python
##now lets get the songlink in the form of text and assign it to songlink_html
songlink_html = response.text
```

```python
#lets see what 'songlink_html gives us'
songlink_html
```

```python
#songlink_html is a mess!!
#lets check if we can extract the lyrics using BeautifulSoup
songlink_soup = BeautifulSoup(songlink_html)
lyric_itself = songlink_soup.body.find(class_ = 'lyric-body').text
lyric_itself
```

```python
type(lyric_itself)
```

```python
#BiautifulSoup is beautiful! but it needs to be cleaned up a little because we have \n \r all over the place
```

```python
# i want to exlude r/ n/ --> they are not words --> here we can seacrh only for the pattern for words


(re.findall('[\w\ ]+', lyric_itself))
```

```python
#and here we can join what we found above!


' '.join(re.findall('[\w\ ]+', lyric_itself)) 
```

```python
# how cool is that!
```

```python
# here now we put it in  a for loop to find ALL the lyrics
```

```python
len(modi_lyric_url)
```

```python
# ask later..

all_beirut_lyrics = []

          
for i in range(len(modi_lyric_url)): 
   # here we can also put 80 in th erange 
    try: 
        songlink = modi_lyric_url[i]
        response =requests.get(songlink)
        songlink_html = response.text
        songlink_soup = BeautifulSoup(songlink_html)
        test = songlink_soup.body.find(class_ = 'lyric-body').text

        all_beirut_lyrics.append(' '.join(re.findall('[\w\t ]+', test))) 
    except: 
        pass
```

```python
all_beirut_lyrics
```

```python
# we have now a list (called all_beirut_lyrics) containing one beirut lyrics in each element 
# the above list will become one of the columns in our dataframe in future 
# the other column inn the dataframe in future must have a lable with name of singer, here beirut, so we need to make another with length similar to the  previous list but 
# only containing 'Beirut'
# lest make that list
```

```python
label = ['Beirut'] * len(all_beirut_lyrics)
print(label)
```

```python
dictx= {'artist': label, 'lyrics': all_beirut_lyrics}
df_Beirut= pd.DataFrame(dictx)
df_Beirut
```

```python
# in this step we have to convert the column lyrics into vectors to be able to use for model
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LogisticRegression
```

```python
our_corpus = list(df['lyrics'])
```

```python
our_corpus
```

```python

vectorizer = TfidfVectorizer(stop_words='english', min_df=10) # note ro self: 10 might be a little high
X = vectorizer.fit_transform(our_corpus)
```

```python

dfX = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names(), index=list(df['artist']))
dfX
```

```python
# next step is defining y for our ML model
# X is already defied (our vectorizer)
y = df['artist']
```

```python
# side note: tokens and their position in feature matrix can be obtained with:
vectorizer.vocabulary_
```

```python
y = list(dfX.index)
print(y)
```

```python
# we do a train/test split but we see that we have much more Decemberists than beirut meaning that we have class impalance 
# we use logistic regression for modeling
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
y_train
```

```python
# and now LogReg
lrc = LogisticRegression().fit(X_train, y_train)
lrc.fit(X_train, y_train)
```

```python
score_train = lrc.score(X_train,y_train)
score_train
```

```python
score_test = lrc.score(X_test,y_test)
score_test
```

```python
# lets test randomforest classifier
rfc = RandomForestClassifier(max_depth = 6)
rfc.fit(X_train, y_train)
```

```python
# lets see what score do we get for train and test data

cv_scores = cross_val_score(rfc, X_train, y_train, cv = 5)
print(f'Train Score: {np.round(np.mean(cv_scores), 3)} +/- {np.round(np.std(cv_scores), 3)}')
```

```python
cv_scores = cross_val_score(rfc, X_test, y_test, cv = 5)
print(f'Test Score: {np.round(np.mean(cv_scores), 3)} +/- {np.round(np.std(cv_scores), 3)}')
```
