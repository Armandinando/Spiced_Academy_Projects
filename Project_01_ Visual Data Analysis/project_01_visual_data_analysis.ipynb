### Project milestone: Animated Scatterplot
#### In this part, the famous Hans Rosling's animation will be reproduced
##### 1. Multiple dataframes will be created using fertility, life expectancy and population. then these dataframes will be transfered into long format with similar number of rows and then will be merged 
##### 2. For each year a separate scatter plot will be created
##### 3. The created scatter plots will be combined to create an animated gif (this method was used to practice data wrangling and visualization. keep in mind that we can also make this animation much faster and easier using the new library called PlotlyExpress that is very simple, fast and easy to use to build Interactive Visualizations.

```python
print(life_expectancy.shape)
print(fertility.shape)
```

```python
life_expectancy.columns
```

```python
fertility.columns
```

```python
# In the fertility dataframe, the years have a type of 'String' while in the life_expectancy, data frame, they are float!
# First, both need to have the same data type!
fertility.columns = fertility.columns.astype(int)
```

```python
fertility.columns
```

```python
# now both have the same data type
```

```python
fertility.index
```

```python
fertility.index.name = 'country'
fertility
```

```python
fertility = fertility.reset_index()
fertility
```

```python
df1 = fertility.melt(id_vars='country', var_name='year', value_name='fertility_rate')
df1
```

```python
#Now we do the same with the other dataframes
life_expectancy
```

```python
life_expectancy.index.name = 'country'
life_expectancy = life_expectancy.reset_index()
df2 = life_expectancy.melt(id_vars='country', var_name='year', value_name='life_expectancy')
df2
```

```python
merged_df = df1.merge(df2)
```

```python
merged_df
```

```python
#A look into population data shows that it has 275 rows. i want it to have the same number of rows as other two dataframes (i. e. 260). 
# therefore i add an argument while reading the excel file (i. e. nrows =260)
```

```python
population = pd.read_excel('/home/arman/github_portfolio_spiced/Project_01_ Visual Data Analysis/gapminder_population.xlsx' , index_col=0, nrows=260)
```

```python
population
```

```python
population.index.name = 'country'
population = population.reset_index()
df3 = population.melt(id_vars='country', var_name='year', value_name='population')
df3
```

```python
merged_df_2= merged_df.merge(df3)
merged_df_2
```

```python
# Now lets's  make a scatter plot for year 2000
df_subset = merged_df_2.loc[merged_df_2['year'] == 2000]
df_subset
```

```python
sns.scatterplot(x='fertility_rate', y= 'life_expectancy',data=df_subset, alpha=0.6)
```

```python
# Now let's try to give it some colors and change the size of dots based on the population
cmap = plt.get_cmap('tab20', lut = len(df_subset)).colors
```

```python
df_subset.plot.scatter('fertility_rate', 'life_expectancy', s=0.8, c=cmap)
```

```python
df_subset.plot.scatter('fertility_rate', 'life_expectancy', s=df_subset['population']*0.000001, c=cmap)
```

```python
# Ok it works! Nice!
```

#### now for preparing an animated scatter plot showing correlation between fertility_rate and life_expectancy (between 1960 to 2015 - before that there are many missing data), we have to create one scatter plot for each year and save them and then glue them together to make a gif!!

```python
#for_1960_2015_df = merged_df_2.loc[(merged_df_2['year'] <= 2015) & (merged_df_2['year'] >= 1960)]
#for_1960_2015_df
```

```python
#create the animation
for i in range(1960,2015):
    for_1960_2015_df = merged_df_2.loc[merged_df_2['year'] == i]
    #plt.figure(figsize = (20,13))
    for_1960_2015_df.plot.scatter('fertility_rate', 'life_expectancy', s=df_subset['population']*0.000001, c=cmap);
    plt.axis([1, 10, 20, 80])
    plt.xlabel('Fertility')
    plt.ylabel('Life Expectancy (Years)')
    plt.title(str(i))
    plt.savefig('graph_'+str(i)+'.png')
    plt.close()
    
images = []
for i in range(1960, 2015):
    filename = 'graph_{}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('Annimated_Plot.gif', images, fps=10)
```
